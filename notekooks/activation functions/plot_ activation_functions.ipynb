```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)

def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    t = tanh(x)
    return 1 - t**2

def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return np.where(x > 0, 1, 0) # 0 for x <= 0, 1 for x > 0

def leaky_relu(x, alpha=0.01):
    return np.maximum(alpha * x, x)

def leaky_relu_derivative(x, alpha=0.01):
    return np.where(x > 0, 1, alpha)

def gelu(x):
    # Approximated version for simplicity
    return 0.5 * x * (1 + np.tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * x**3)))

def gelu_derivative(x):
    # Numerical approximation for simplicity in plotting.
    # Analytical derivative is more complex.
    delta = 1e-5
    return (gelu(x + delta) - gelu(x - delta)) / (2 * delta)

# Generate a range of x values
x = np.linspace(-5, 5, 400)

plt.figure(figsize=(15, 10))

# Plot Sigmoid
plt.subplot(2, 3, 1)
plt.plot(x, sigmoid(x))
plt.title('Sigmoid Activation')
plt.grid(True)

plt.subplot(2, 3, 4)
plt.plot(x, sigmoid_derivative(x))
plt.title('Sigmoid Derivative')
plt.grid(True)

# Plot Tanh
plt.subplot(2, 3, 2)
plt.plot(x, tanh(x))
plt.title('Tanh Activation')
plt.grid(True)

plt.subplot(2, 3, 5)
plt.plot(x, tanh_derivative(x))
plt.title('Tanh Derivative')
plt.grid(True)

# Plot ReLU
plt.subplot(2, 3, 3)
plt.plot(x, relu(x))
plt.title('ReLU Activation')
plt.grid(True)

plt.subplot(2, 3, 6)
plt.plot(x, relu_derivative(x))
plt.title('ReLU Derivative')
plt.grid(True)

plt.tight_layout()
plt.show()

# Plot Leaky ReLU and GELU separately for clarity
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(x, leaky_relu(x))
plt.title('Leaky ReLU Activation')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x, gelu(x))
plt.title('GELU Activation')
plt.grid(True)

plt.tight_layout()
plt.show()

# And their derivatives
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(x, leaky_relu_derivative(x))
plt.title('Leaky ReLU Derivative')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x, gelu_derivative(x))
plt.title('GELU Derivative (Numerical)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erf

# Gerar um range de valores x
x = np.linspace(-5, 5, 1000)

# 1. Sigmoid
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
def d_sigmoid(x):
    s = sigmoid(x)
    return s * (1 - s)

# 2. Tanh
def tanh(x):
    return np.tanh(x)
def d_tanh(x):
    return 1 - np.tanh(x)**2

# 3. ReLU
def relu(x):
    return np.maximum(0, x)
def d_relu(x):
    return np.where(x > 0, 1, 0) # Derivada é 1 para x>0, 0 para x<=0

# 4. Leaky ReLU
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)
def d_leaky_relu(x, alpha=0.01):
    return np.where(x > 0, 1, alpha)

# 5. GELU (aproximação para visualização simples)
# A implementação exata de GELU envolve a função de erro (erf)
# Para visualização, podemos usar uma aproximação comum ou a implementação de PyTorch/TensorFlow
# Aqui, usaremos a implementação direta com scipy.special.erf para precisão.
def gelu(x):
    return 0.5 * x * (1 + erf(x / np.sqrt(2)))
# A derivada de GELU é mais complexa, mas podemos plotar a função para ver a suavidade.
# d/dx (x * Phi(x)) = Phi(x) + x * phi(x) onde phi é a PDF da normal padrão
# phi(x) = (1/sqrt(2*pi)) * exp(-x^2/2)
def d_gelu(x):
    phi_x = (1 / np.sqrt(2 * np.pi)) * np.exp(-x**2 / 2)
    Phi_x = 0.5 * (1 + erf(x / np.sqrt(2)))
    return Phi_x + x * phi_x

# Plotar as funções de ativação
plt.figure(figsize=(12, 8))
plt.plot(x, sigmoid(x), label='Sigmoid')
plt.plot(x, tanh(x), label='Tanh')
plt.plot(x, relu(x), label='ReLU')
plt.plot(x, leaky_relu(x), label='Leaky ReLU ($\\alpha=0.01$)')
plt.plot(x, gelu(x), label='GELU')
plt.title('Funções de Ativação')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.axvline(0, color='gray', linestyle='--', linewidth=0.8)
plt.axhline(0, color='gray', linestyle='--', linewidth=0.8)
plt.show()

# Plotar as derivadas das funções de ativação
plt.figure(figsize=(12, 8))
plt.plot(x, d_sigmoid(x), label='Derivada da Sigmoid')
plt.plot(x, d_tanh(x), label='Derivada da Tanh')
plt.plot(x, d_relu(x), label='Derivada da ReLU')
plt.plot(x, d_leaky_relu(x), label='Derivada da Leaky ReLU')
plt.plot(x, d_gelu(x), label='Derivada da GELU') # Adicionando a derivada da GELU
plt.title('Derivadas das Funções de Ativação')
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.grid(True)
plt.legend()
plt.axvline(0, color='gray', linestyle='--', linewidth=0.8)
plt.axhline(0, color='gray', linestyle='--', linewidth=0.8)
plt.show()
```
