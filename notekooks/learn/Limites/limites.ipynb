## Limites

Excelente! O próximo passo no nosso roadmap é a **Introdução a Limites**.

Os limites são um conceito absolutamente fundamental no cálculo. Eles formam a base para entender a continuidade das funções e, mais importante para o nosso objetivo, para definir o que é uma derivada.

### 1. O que é um Limite? (Noção Intuitiva)

Imagine que você está investigando o comportamento de uma função $f(x)$ à medida que o valor de $x$ se aproxima *muito, muito perto* de um certo número, digamos, $p$. Você quer saber para qual valor a saída $f(x)$ está se aproximando. Esse valor para o qual $f(x)$ se aproxima é o **limite** da função $f(x)$ quando $x$ tende a $p$.

* Informalmente, dizemos que uma função $f$ tem um limite $L$ em uma entrada $p$, se $f(x)$ fica cada vez mais perto de $L$ à medida que $x$ se aproxima cada vez mais de $p$[cite: 3].
* O interessante é que o comportamento da função *exatamente no ponto* $p$ (ou seja, o valor de $f(p)$) não importa para a definição do limite. A função pode nem estar definida no ponto $p$[cite: 1, 31]. Estamos interessados no que acontece *ao redor* de $p$.

**Analogia da Paisagem (adaptado de [cite: 16, 17, 18, 19, 20]):**
Imagine que você está caminhando em uma paisagem representada pelo gráfico de $y = f(x)$. Sua posição horizontal é $x$ e sua altitude é $y$. Você caminha em direção a uma posição horizontal $x=p$. À medida que você se aproxima cada vez mais de $p$ (sem necessariamente chegar em $p$), você observa que sua altitude se aproxima cada vez mais de um valor específico $L$. Esse valor $L$ seria o limite da sua altitude quando sua posição horizontal se aproxima de $p$.

A ideia chave é que podemos fazer o valor da função $f(x)$ ficar arbitrariamente próximo do limite $L$, simplesmente escolhendo valores de $x$ que sejam suficientemente próximos de $p$ (mas diferentes de $p$)[cite: 4, 22].

### 2. Por que os Limites são Importantes?

* **Base para as Derivadas:** A definição formal de uma derivada (que veremos a seguir) é dada por um limite. A derivada mede a taxa instantânea de variação de uma função, e essa "instantaneidade" é capturada através de um processo de limite.
* **Continuidade:** Os limites ajudam a definir formalmente se uma função é contínua em um ponto (ou seja, se o gráfico não tem "quebras" ou "saltos"). Uma função é contínua em um ponto se o limite nesse ponto existe, a função está definida nesse ponto, e o limite é igual ao valor da função nesse ponto[cite: 7].
* **Comportamento Assintótico:** Limites também são usados para descrever o comportamento de funções quando $x$ tende ao infinito (ou menos infinito), o que nos ajuda a entender as assíntotas dos gráficos.

### 3. Exemplo Clássico: O Limite de $\frac{\sin x}{x}$ quando $x \to 0$

Considere a função $f(x) = \frac{\sin x}{x}$.
* Esta função não está definida em $x=0$, porque teríamos uma divisão por zero.
* No entanto, podemos investigar o que acontece com $f(x)$ quando $x$ se aproxima *muito* de 0 (sem ser 0).

Se você calcular os valores de $f(x)$ para $x$ cada vez mais próximos de 0 (por exemplo, $x=0.1$, $x=0.01$, $x=0.001$, e também $x=-0.1$, $x=-0.01$, $x=-0.001$), você verá que os valores de $f(x)$ se aproximam cada vez mais de 1.
* A tabela no documento fornecido mostra isso:
    * Para $x=1$, $\frac{\sin x}{x} \approx 0.841471...$
    * Para $x=0.1$, $\frac{\sin x}{x} \approx 0.998334...$
    * Para $x=0.01$, $\frac{\sin x}{x} \approx 0.999983...$ [cite: 10]
* Portanto, dizemos que o limite de $\frac{\sin x}{x}$ quando $x$ tende a 0 é igual a 1[cite: 11]. Escrevemos isso como:
    $$\lim_{x\rightarrow 0} \frac{\sin x}{x} = 1$$

Isso ilustra perfeitamente que o limite pode existir mesmo que a função não esteja definida no ponto exato[cite: 11, 31].

### 4. Ideia de Proximidade (A Definição Formal $(\epsilon, \delta)$ em Poucas Palavras)

A noção intuitiva de "se aproximar" é formalizada na matemática pela definição $(\epsilon, \delta)$ (épsilon-delta) de um limite[cite: 2, 23]. Você não precisa dominar essa definição formal em profundidade para o backpropagation agora, mas é bom saber que ela existe.

Basicamente, a definição $(\epsilon, \delta)$ diz:
O limite de $f(x)$ quando $x$ se aproxima de $p$ é $L$ se, para qualquer pequena margem de erro $\epsilon > 0$ que você escolher em torno de $L$, existe uma "zona de proximidade" $\delta > 0$ ao redor de $p$ tal que, se $x$ estiver dentro dessa zona de proximidade de $p$ (mas $x \neq p$), então $f(x)$ estará dentro da margem de erro $\epsilon$ de $L$. [cite: 24]

Simbolicamente:
$(\forall\epsilon>0)(\exists\delta>0)(\forall x\in\mathbb{R})(0<|x-p|<\delta \Rightarrow |f(x)-L|<\epsilon)$ [cite: 24]

Isso apenas formaliza a ideia de que podemos fazer $f(x)$ tão próximo de $L$ quanto quisermos, tornando $x$ suficientemente próximo de $p$[cite: 22].

### 5. Quando um Limite Pode Não Existir?

Um limite pode não existir em algumas situações[cite: 5]:
* **Saltos:** Se a função se aproxima de valores diferentes quando $x$ se aproxima de $p$ pela esquerda versus pela direita (limites laterais diferentes)[cite: 39, 40].
* **Comportamento Oscilatório Infinito:** Se a função oscila demais perto de $p$ sem se estabilizar em torno de um valor.
* **Crescimento Ilimitado (Assíntotas Verticais):** Se $f(x)$ cresce (ou decresce) indefinidamente quando $x$ se aproxima de $p$. (Ex: $\lim_{x\rightarrow 0} \frac{1}{x^2} = \infty$).

**Como Explorar Limites com Python (Sugestão):**

Você pode usar Python para "ver" um limite numericamente, assim como fizemos com o exemplo de $\frac{\sin x}{x}$:


```python
import numpy as np

def f(x):
    if x == 0:
        return np.nan # Indefinido em x=0
    return np.sin(x) / x

# Valores de x se aproximando de 0
valores_x_positivos = np.array([0.1, 0.01, 0.001, 0.0001])
valores_x_negativos = np.array([-0.1, -0.01, -0.001, -0.0001])

print("f(x) para x > 0 se aproximando de 0:")
for x_val in valores_x_positivos:
    print(f"f({x_val:.4f}) = {f(x_val):.6f}")

print("\nf(x) para x < 0 se aproximando de 0:")
for x_val in valores_x_negativos:
    print(f"f({x_val:.4f}) = {f(x_val):.6f}")

```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Definição da função f(x) = sin(x)/x
def f(x_input):
    """
    Calcula sin(x)/x. Retorna np.nan se x for 0.
    Aceita tanto escalares quanto arrays NumPy como entrada.
    """
    if isinstance(x_input, (int, float)): # Se a entrada for um único número
        if np.isclose(x_input, 0.0):
            return np.nan
        else:
            return np.sin(x_input) / x_input
    else: # Se a entrada for um array NumPy
        # Criar um array para os resultados
        resultado = np.zeros_like(x_input, dtype=float)
        # Calcular sin(x)/x para x != 0
        mascara_nao_zero = ~np.isclose(x_input, 0.0)
        valores_nao_zero = x_input[mascara_nao_zero]
        resultado[mascara_nao_zero] = np.sin(valores_nao_zero) / valores_nao_zero
        # Definir como np.nan para x = 0
        resultado[np.isclose(x_input, 0.0)] = np.nan
        return resultado

# 2. Gerar valores de x para o plot
# Intervalo simétrico para mostrar o comportamento da função
x_plot = np.linspace(-2 * np.pi, 2 * np.pi, 500) # 501 pontos para incluir 0.0 exatamente

# 3. Calcular os valores de y
# A função f lida com x=0 retornando np.nan, que o matplotlib trata como uma quebra.
y_plot = f(x_plot)

# 4. O limite conhecido da função quando x -> 0 é 1
limite_y_conhecido = 1.0

# 5. Criar o gráfico
plt.figure(figsize=(10, 6)) # Tamanho da figura um pouco maior

# Plotar a função f(x)
plt.plot(x_plot, y_plot, label="$f(x) = \\frac{\\sin x}{x}$", color="dodgerblue", linewidth=2)

# Marcar o limite em (0,1) com um círculo aberto (oco)
# Isso indica que a função "tende" para este ponto.
plt.plot(0, limite_y_conhecido,
         'o',                          # Tipo de marcador: círculo
         markersize=8,                 # Tamanho do marcador
         markerfacecolor='white',      # Cor de preenchimento do marcador
         markeredgecolor='red',        # Cor da borda do marcador
         markeredgewidth=1.5,          # Largura da borda
         label=f"Limite em $x=0$ é ${limite_y_conhecido}$")

# 6. Melhorar a estética e informação do gráfico
plt.title("Comportamento de $f(x) = \\frac{\\sin x}{x}$ perto de $x=0$", fontsize=15)
plt.xlabel("$x$", fontsize=12)
plt.ylabel("$f(x)$", fontsize=12)

# Adicionar linhas dos eixos x=0 e y=0 para referência
plt.axhline(0, color='black', linewidth=0.7, linestyle='--')
plt.axvline(0, color='black', linewidth=0.7, linestyle='--')

# Definir limites dos eixos para melhor visualização, se necessário
# plt.ylim(-0.5, 1.2) # Exemplo de ajuste do eixo y
# plt.xlim(x_plot.min(), x_plot.max())

plt.legend(fontsize=10)
plt.grid(True, linestyle=':', alpha=0.7) # Grid mais suave

plt.show()
```
