## Fundamentos de Funções

**O que é uma função?**

Entender a relação entre entrada (input) e saída (output). 

**Tipos de funções**: 
- Lineares, 
- Polinomiais, 
- Exponenciais, 
- Logarítmicas, 
- Trigonométricas (especialmente seno e cosseno, que aparecem em exemplos).

Composição de funções: Entender o que é f(g(x)). Isso é a base da regra da cadeia.  

Por exemplo, se g(x)=x^2 e f(x)=sin(x), então f(g(x))=sin(x^2). 



Ótimo! Começar pelos fundamentos das funções é o passo certo. Uma função é um dos blocos de construção mais essenciais da matemática, especialmente para o cálculo e, por extensão, para o backpropagation.

Vamos detalhar os principais aspectos:

### 1. O que é uma Função?

Imagine uma máquina que recebe uma **entrada** (input) e, seguindo uma regra específica, produz uma única **saída** (output) correspondente. Essa "máquina" é uma função.

* **Definição Informal:** Uma função `f` associa uma saída `f(x)` a cada entrada `x`[cite: 3]. A função pode ou não estar definida para todas as entradas possíveis.
* **Entrada (Input ou Argumento):** É o valor que "alimentamos" na função. Geralmente representado por `x` ou outra variável.
* **Saída (Output ou Valor da Função):** É o valor que a função produz para uma determinada entrada. Geralmente representado por `f(x)` ou `y`.
* **Regra:** É a operação ou conjunto de operações que a função aplica à entrada para gerar a saída. Por exemplo, na função $f(x) = x^2$ (a função de elevar ao quadrado [cite: 119]), a regra é "elevar a entrada ao quadrado".
* **Domínio (Domain):** É o conjunto de todas as entradas possíveis para as quais a função está definida[cite: 112]. Por exemplo, na função $f(x) = \frac{1}{x}$, o domínio são todos os números reais, exceto o zero (pois a divisão por zero não é definida). Na função $f(x) = \sqrt{x}$, o domínio (considerando números reais) são todos os números reais não negativos ($x \ge 0$).
* **Contradomínio (Codomain):** É o conjunto de todas as saídas *potenciais* da função.
* **Imagem (Image/Range):** É o conjunto de todas as saídas *reais* que a função produz a partir do seu domínio. A imagem é um subconjunto do contradomínio.

**Notação Comum:**
Usamos a notação $f: A \rightarrow B$, que significa que `f` é uma função que mapeia elementos do conjunto `A` (domínio) para elementos do conjunto `B` (contradomínio). E escrevemos $y = f(x)$ para indicar que `y` é a saída da função `f` quando a entrada é `x`.

**Exemplo Simples:**
Seja a função $f(x) = 2x + 1$.
* Se a entrada $x = 3$, a saída $f(3) = 2(3) + 1 = 7$.
* O domínio poderia ser todos os números reais ($\mathbb{R}$).
* A imagem também seria todos os números reais ($\mathbb{R}$).

### 2. Tipos Comuns de Funções

Existem muitos tipos de funções. Algumas das mais relevantes para o estudo de cálculo e redes neurais incluem:

* **Funções Lineares:** Têm a forma $f(x) = mx + c$. Seu gráfico é uma reta.
    * Exemplo: $f(x) = 3x - 2$.
* **Funções Polinomiais:** Envolvem potências de $x$.
    * Exemplo: $f(x) = x^2$ (função quadrática)[cite: 119], $f(x) = x^3 - 2x + 5$ (função cúbica).
* **Funções Exponenciais:** A variável está no expoente.
    * Exemplo: $f(x) = e^x$ (onde `e` é a base do logaritmo natural, aproximadamente 2.71828 [cite: 168]), $f(x) = a^x$[cite: 169].
* **Funções Logarítmicas:** São as inversas das funções exponenciais.
    * Exemplo: $f(x) = \ln(x)$ (logaritmo natural, base `e`)[cite: 169], $f(x) = \log_a(x)$[cite: 169].
* **Funções Trigonométricas:** Relacionam ângulos com as razões dos lados de um triângulo retângulo.
    * Exemplo: $f(x) = \sin(x)$[cite: 169], $f(x) = \cos(x)$[cite: 170], $f(x) = \tan(x)$[cite: 170]. A função $\frac{\sin x}{x}$ é um exemplo clássico em limites, embora não definida em $x=0$, seu limite quando $x$ se aproxima de zero é 1[cite: 11].
* **Funções de Valor Absoluto:** $f(x) = |x|$. Retorna o valor não negativo de $x$. Esta função é contínua em $x=0$, mas não é diferenciável nesse ponto[cite: 139, 140].
* **Funções Racionais:** São razões de dois polinômios, como $f(x) = \frac{p(x)}{q(x)}$[cite: 77].
    * Exemplo: $f(x) = \frac{1}{x}$.
* **Funções Definidas por Partes (Piecewise Functions):** Funções definidas por diferentes regras em diferentes partes do seu domínio.
    * Exemplo: A função degrau (step function) que retorna 1 para $x < a$ e 10 para $x \ge a$[cite: 132].

### 3. Composição de Funções

A composição de funções é uma ideia central e absolutamente crucial para entender a regra da cadeia. É como aplicar uma função ao resultado de outra função.

* **Definição:** Se temos duas funções, $f$ e $g$, a composição de $f$ com $g$, denotada por $f \circ g$ (lê-se "f círculo g" ou "f composta com g"), é definida como:
    $(f \circ g)(x) = f(g(x))$ [cite: 291]
    Isso significa: primeiro aplicamos a função $g$ a $x$, obtemos $g(x)$, e então aplicamos a função $f$ a esse resultado $g(x)$.

* **Ordem Importa:** Em geral, $f(g(x)) \neq g(f(x))$.

* **Exemplo 1:**
    Seja $g(x) = x^2$ e $f(u) = \sin(u)$.
    Então, $(f \circ g)(x) = f(g(x)) = f(x^2) = \sin(x^2)$[cite: 314].
    E $(g \circ f)(x) = g(f(x)) = g(\sin(x)) = (\sin(x))^2 = \sin^2(x)$.

* **Exemplo 2 (com três funções):**
    Seja $h(x) = x^2$, $g(v) = \sin(v)$, e $f(u) = e^u$.
    Queremos encontrar $y = f(g(h(x)))$.
    1.  $v = h(x) = x^2$
    2.  $u = g(v) = g(x^2) = \sin(x^2)$
    3.  $y = f(u) = f(\sin(x^2)) = e^{\sin(x^2)}$ [cite: 314]

Entender a composição é o primeiro grande passo para dominar a regra da cadeia, pois a regra da cadeia nos ensina como derivar funções compostas.

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_function(func, x_range, title, label_func, num_points=400, color=None, linestyle='-'):
    """
    Plota uma dada função sobre um intervalo especificado de forma elegante.

    Args:
        func (callable): A função a ser plotada. Deve aceitar um array numpy como entrada.
        x_range (tuple): Uma tupla (min, max) para o eixo x.
        title (str): O título do gráfico.
        label_func (str): O rótulo para a legenda, preferencialmente em formato LaTeX.
        num_points (int): Número de pontos para gerar no intervalo x para a curva.
        color (str, optional): Cor da linha do gráfico.
        linestyle (str, optional): Estilo da linha do gráfico.
    """
    x_values = np.linspace(x_range[0], x_range[1], num_points)
    y_values = func(x_values)

    plt.figure(figsize=(8, 6))
    plt.plot(x_values, y_values, label=label_func, color=color, linestyle=linestyle)
    plt.title(title, fontsize=15)
    plt.xlabel("Eixo $x$", fontsize=12)
    plt.ylabel("Eixo $f(x)$", fontsize=12)

    # Adiciona linhas para os eixos x e y cruzando em (0,0)
    plt.axhline(0, color='black', linewidth=0.7)
    plt.axvline(0, color='black', linewidth=0.7)

    # Adiciona um grid para melhor leitura
    plt.grid(True, linestyle=':', alpha=0.6)

    plt.legend(fontsize=12)
    plt.show()

def plot_function_on_ax(ax, func, x_range, title, label_func, num_points=200, color=None, linestyle='-'):
    """
    Plota uma dada função em um objeto Axes (subplot) do Matplotlib.

    Args:
        ax (matplotlib.axes.Axes): O eixo (subplot) onde a função será plotada.
        func (callable): A função a ser plotada. Deve aceitar um array numpy como entrada.
        x_range (tuple): Uma tupla (min, max) para o eixo x.
        title (str): O título do subplot.
        label_func (str): O rótulo para a legenda, preferencialmente em formato LaTeX.
        num_points (int): Número de pontos para gerar no intervalo x para a curva.
        color (str, optional): Cor da linha do gráfico.
        linestyle (str, optional): Estilo da linha do gráfico.
    """
    x_values = np.linspace(x_range[0], x_range[1], num_points)
    y_values = func(x_values)

    ax.plot(x_values, y_values, label=label_func, color=color, linestyle=linestyle)
    ax.set_title(title, fontsize=10)  # Fonte menor para subplots
    ax.set_xlabel("Eixo $x$", fontsize=8)
    ax.set_ylabel("Eixo $f(x)$", fontsize=8)

    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)
    ax.grid(True, linestyle=':', alpha=0.5)
    ax.legend(fontsize=7) # Fonte da legenda menor
    ax.tick_params(axis='both', which='major', labelsize=7) # Tamanho dos ticks

# --- Definições das Funções ---

def func_linear(x_input):
    """Função linear: f(x) = 3x - 2"""
    return 3 * x_input - 2

def func_quadratica(x_input):
    """Função quadrática: f(x) = x^2"""
    return x_input**2

def func_polinomial_cubica(x_input):
    """Função polinomial cúbica: f(x) = x^3 - 2x + 5"""
    return x_input**3 - 2 * x_input + 5

def func_exponencial(x_input):
    """Função exponencial: f(x) = e^x"""
    return np.exp(x_input)

def func_logaritmica(x_input):
    """Função logarítmica natural: f(x) = ln(x)"""
    # Adiciona um pequeno epsilon para evitar log(0) ou valores negativos se o range incluir <=0
    # No entanto, o ideal é garantir que x_range[0] > 0 para ln(x)
    return np.log(x_input)

def func_seno(x_input):
    """Função seno: f(x) = sin(x)"""
    return np.sin(x_input)

def func_valor_absoluto(x_input):
    """
    Função de Valor Absoluto: f(x) = |x|.
    Retorna o valor não negativo de x.
    Esta função é contínua em x=0, mas não é diferenciável nesse ponto.
    """
    return np.abs(x_input)

def func_racional_exemplo(x_input):
    """
    Função Racional Exemplo: f(x) = 1/x.
    São razões de dois polinômios.
    """
    try:

        resultado = 1.0 / x_input
        if isinstance(x_input, np.ndarray):
            resultado[x_input == 0] = np.nan
        return resultado

    except Exception as e:
        return np.nan # Retorna NaN para entrada escalar zero

def func_degrau(x_input, a=0):
    """
    Função Degrau (Step Function) - Exemplo de Função Definida por Partes.
    Retorna 1 para x < a e 10 para x >= a.
    """
    return np.where(x_input < a, 1, 10)


# --- Plotando as Funções ---
# --- Criando a Figura e a Grade de Subplots ---
# Grid 3x3 para acomodar até 9 gráficos. Usaremos 6.
fig, axes = plt.subplots(3, 3, figsize=(15, 15))

# Achatando a matriz de eixos (de 3x3 para um array de 9) para facilitar a iteração
axes_flat = axes.flatten()

# Lista de dicionários, cada um contendo os parâmetros para plotar uma função
functions_to_plot = [
    {'func': func_linear, 'x_range': (-5, 5), 'title': "Linear", 'label': "$f(x) = 3x - 2$", 'color': 'blue'},
    {'func': func_quadratica, 'x_range': (-5, 5), 'title': "Quadrática", 'label': "$f(x) = x^2$", 'color': 'orange'},
    {'func': func_polinomial_cubica, 'x_range': (-4, 4), 'title': "Cúbica", 'label': "$f(x) = x^3 - 2x + 5$", 'color': 'green'},
    {'func': func_exponencial, 'x_range': (-2, 3), 'title': "Exponencial", 'label': "$f(x) = e^x$", 'color': 'red'},
    {'func': func_logaritmica, 'x_range': (0.1, 10), 'title': "Logarítmica", 'label': "$f(x) = \\ln(x)$", 'color': 'purple'},
    {'func': func_seno, 'x_range': (-2 * np.pi, 2 * np.pi), 'title': "Seno", 'label': "$f(x) = \\sin(x)$", 'color': 'magenta'},
    {'func': func_degrau, 'x_range': (-5, 10), 'title': "Degral", 'label': "", 'color': 'magenta'},
    {'func': func_racional_exemplo, 'x_range': (-5, 10), 'title': "Racional", 'label': "", 'color': 'magenta'},
    {'func': func_valor_absoluto, 'x_range': (-5, 10), 'title': "Valor Absoluto", 'label': "", 'color': 'magenta'},

]

# Plotar cada função em seu respectivo subplot
for i, plot_params in enumerate(functions_to_plot):
    plot_function_on_ax(axes_flat[i],
                        plot_params['func'],
                        plot_params['x_range'],
                        plot_params['title'],
                        plot_params['label'],
                        color=plot_params['color'])

# Ocultar os eixos não utilizados (se houver mais eixos do que funções)
for i in range(len(functions_to_plot), len(axes_flat)):
    axes_flat[i].axis('off')

# Adicionar um título geral para toda a figura
plt.suptitle("Galeria de Funções Fundamentais", fontsize=16, y=0.99) # y ajustado para não sobrepor

# Ajustar o layout para evitar sobreposição de títulos/rótulos
plt.tight_layout(rect=[0, 0, 1, 0.97]) # rect ajustado para dar espaço ao suptitle

plt.show()
```

### Composição de Funções

Você captou a essência da mecânica da composição de funções: **aplicar uma função ao resultado de outra**. No entanto, o foco principal da composição de funções não é necessariamente a "simplificação" no sentido de tornar a expressão mais curta ou mais fácil de calcular manualmente, embora isso possa acontecer em casos específicos.

Vamos refinar o seu entendimento:

**O Principal Objetivo da Composição de Funções ($f(g(x))$):**

1.  **Modelar Processos em Etapas:** A composição é fundamental para descrever situações onde um processo ocorre em etapas sequenciais. A saída de uma etapa (função $g$) se torna a entrada para a próxima etapa (função $f$).
    * **Exemplo do seu texto:** Para $y = e^{\sin(x^2)}$, podemos pensar nisso como:
        1.  Etapa 1 (função $h$): Elevar $x$ ao quadrado ($v = x^2$).
        2.  Etapa 2 (função $g$): Calcular o seno do resultado anterior ($u = \sin(v) = \sin(x^2)$).
        3.  Etapa 3 (função $f$): Calcular $e$ elevado ao resultado anterior ($y = e^u = e^{\sin(x^2)}$).

2.  **Construir Funções Mais Complexas a Partir de Funções Simples:** Muitas funções complexas podem ser entendidas como composições de funções mais básicas e conhecidas. Isso ajuda a analisar e entender o comportamento dessas funções complexas.

3.  **Preparação para a Regra da Cadeia (Chain Rule) no Cálculo:** Este é o ponto crucial para o seu estudo de backpropagation. A composição de funções é a estrutura sobre a qual a regra da cadeia opera. A regra da cadeia nos ensina como encontrar a derivada de uma função composta ($f(g(x))$) utilizando as derivadas das funções individuais ($f$ e $g$).

**Em Resumo:**

* Quando você tem $f(x)$ e $g(x)$, e $f$ é composta por $g$ (resultando em $f(g(x))$), você está definindo uma nova função que representa a aplicação sequencial de $g$ e depois $f$.
* O objetivo não é "achar uma simplificação" da *expressão* $f(g(x))$ em todos os casos, mas sim:
    * **Entender a estrutura da função composta.**
    * **Modelar processos em etapas.**
    * **Preparar o terreno para aplicar a regra da cadeia**, que é essencial para calcular derivadas de tais funções – algo vital no backpropagation para entender como o erro se propaga através das "camadas" (funções) de uma rede neural.

Seu entendimento da mecânica de $f(g(x))$ está correto. Apenas expanda a visão para além da "simplificação" e foque na ideia de construir e analisar funções em etapas, o que é a base para a regra da cadeia.

```python
import numpy as np # Usaremos numpy para a função seno (np.sin) que lida bem com números e arrays

# 1. Definir as funções individuais
def g(x):
    """Calcula x ao quadrado."""
    print(f"  Dentro de g(x): recebido x = {x}, retornando x^2 = {x**2}")
    return x**2

def f(u):
    """Calcula o seno de u."""
    print(f"  Dentro de f(u): recebido u = {u}, retornando sin(u) = {np.sin(u):.4f}")
    return np.sin(u)

# 2. Definir as composições

# Composição (f O g)(x) = f(g(x))
def f_composta_g(x_input):
    """Calcula f(g(x_input))."""
    print(f"Calculando f(g({x_input})):")
    resultado_g = g(x_input)      # Primeiro aplica g
    resultado_final = f(resultado_g) # Depois aplica f ao resultado de g
    print(f"Resultado final de f(g({x_input})) = {resultado_final:.4f}\n")
    return resultado_final

# Composição (g O f)(x) = g(f(x))
def g_composta_f(x_input):
    """Calcula g(f(x_input))."""
    print(f"Calculando g(f({x_input})):")
    resultado_f = f(x_input)      # Primeiro aplica f
    resultado_final = g(resultado_f) # Depois aplica g ao resultado de f
    print(f"Resultado final de g(f({x_input})) = {resultado_final:.4f}\n")
    return resultado_final

# 3. Testar as composições
# valor_exemplo = 2
# resultado_f_g = f_composta_g(valor_exemplo)
# resultado_g_f = g_composta_f(valor_exemplo)


plt.plot([f_composta_g(x) for x in np.linspace(-5, 5, 10)])
```

```python
from sympy import symbols, diff
x, y = symbols('x y')
f = x**2 + y**2
gradiente = [diff(f, var) for var in (x, y)]
print(gradiente)
```
