## Derivadas

Excelente! Avançando no nosso roadmap, chegamos a um dos conceitos mais importantes do cálculo: **A Derivada**.

A derivada nos permite quantificar como uma função muda à medida que sua entrada varia. É uma ferramenta fundamental para entender taxas de variação, otimização (encontrar máximos e mínimos) e muitos outros problemas em ciência e engenharia, incluindo o funcionamento interno do backpropagation.

### 1. O que é a Derivada? Duas Perspectivas Intuitivas

Podemos entender a derivada de duas formas principais:

* **Taxa de Variação Instantânea:** A derivada de uma função em um ponto nos diz quão rapidamente o valor da função está mudando *exatamente naquele ponto*[cite: 102]. Pense na velocidade de um carro: o velocímetro mostra a velocidade instantânea, que é a derivada da posição do carro em relação ao tempo.
* **Inclinação da Reta Tangente:** Geometricamente, a derivada de uma função $f(x)$ em um ponto $x=a$ é a inclinação da reta que "apenas toca" o gráfico da função $f(x)$ nesse ponto $a$[cite: 97, 99]. Essa reta é chamada de reta tangente e representa a melhor aproximação linear da função perto daquele ponto[cite: 98].

Ambas as perspectivas são cruciais. A inclinação da tangente nos dá uma representação visual da taxa de variação instantânea.

### 2. Definição Formal da Derivada (Usando Limites)

A intuição é ótima, mas a matemática precisa de uma definição precisa. A derivada é formalmente definida usando o conceito de limites que acabamos de discutir.

Seja uma função $f(x)$ e um ponto $a$ no seu domínio. A derivada de $f$ em $a$, denotada por $f'(a)$ (lê-se "f linha de a"), é definida como:

$$f'(a) = \lim_{h\rightarrow0} \frac{f(a+h) - f(a)}{h}$$

Desde que este limite exista[cite: 112].

**Vamos entender essa fórmula:**

* $f(a)$: O valor da função no ponto $a$.
* $f(a+h)$: O valor da função em um ponto um pouquinho distante de $a$ (a uma distância $h$).
* $f(a+h) - f(a)$: A mudança no valor da função (a variação em $y$, ou $\Delta y$) quando $x$ varia de $a$ para $a+h$.
* $h$: A mudança no valor de $x$ (a variação em $x$, ou $\Delta x$).
* $\frac{f(a+h) - f(a)}{h}$: Esta é a inclinação da reta secante que passa pelos pontos $(a, f(a))$ e $(a+h, f(a+h))$ no gráfico da função[cite: 121]. É a taxa de variação *média* da função no intervalo de $a$ até $a+h$.
* $\lim_{h\rightarrow0}$: À medida que $h$ (a distância entre os dois pontos) se torna cada vez menor, aproximando-se de zero, a reta secante se aproxima da reta tangente. O limite dessa inclinação da reta secante, se existir, nos dá a inclinação da reta tangente no ponto $a$, que é a taxa de variação *instantânea*[cite: 122].

Essa definição usando limites é um exemplo da definição $(\epsilon, \delta)$ que vimos anteriormente[cite: 115].

### 3. Notações Comuns para a Derivada

Existem várias maneiras de denotar a derivada, dependendo do contexto e da preferência:

* **Notação de Lagrange:** $f'(a)$ (lê-se "f linha de a") para a derivada em um ponto específico $a$, ou $f'(x)$ para a função derivada (que veremos a seguir)[cite: 117]. Se $y=f(x)$, também podemos escrever $y'$.
* **Notação de Leibniz:** $\frac{df}{dx}(a)$ (lê-se "a derivada de f em relação a x no ponto a") ou, se $y=f(x)$, $\frac{dy}{dx}\Big|_{x=a}$[cite: 117]. Esta notação é muito útil pois enfatiza que a derivada é uma taxa de variação de $f$ (ou $y$) em relação a $x$. Para a função derivada, escrevemos $\frac{dy}{dx}$ ou $\frac{df}{dx}$[cite: 151].

### 4. Exemplo Prático: Derivada de $f(x) = x^2$

Vamos usar a definição formal para encontrar a derivada da função $f(x) = x^2$ em um ponto $a$.

$f'(a) = \lim_{h\rightarrow0} \frac{f(a+h) - f(a)}{h}$
$f'(a) = \lim_{h\rightarrow0} \frac{(a+h)^2 - a^2}{h}$ [cite: 119]
$f'(a) = \lim_{h\rightarrow0} \frac{(a^2 + 2ah + h^2) - a^2}{h}$ [cite: 119]
$f'(a) = \lim_{h\rightarrow0} \frac{2ah + h^2}{h}$
$f'(a) = \lim_{h\rightarrow0} \frac{h(2a + h)}{h}$

Como $h \rightarrow 0$ mas $h \neq 0$, podemos cancelar $h$ no numerador e denominador:
$f'(a) = \lim_{h\rightarrow0} (2a + h)$
Agora, substituindo $h=0$:
$f'(a) = 2a + 0 = 2a$

Portanto, a derivada da função $f(x) = x^2$ no ponto $a$ é $2a$[cite: 120]. Por exemplo, no ponto $x=3$, a inclinação da tangente ao gráfico de $f(x)=x^2$ é $2(3) = 6$.

### 5. A Função Derivada

Se uma função $f$ tem uma derivada em cada ponto $x$ do seu domínio (ou em um subconjunto dele), podemos definir uma nova função, chamada **função derivada** (ou simplesmente a derivada de $f$), denotada por $f'(x)$ (ou $\frac{dy}{dx}$)[cite: 118]. Esta nova função associa a cada ponto $x$ o valor da derivada de $f$ naquele ponto.

Para o nosso exemplo $f(x) = x^2$, a função derivada é $f'(x) = 2x$[cite: 120].

### 6. Relação entre Diferenciabilidade e Continuidade

* **Diferenciabilidade implica Continuidade:** Se uma função $f$ é diferenciável em um ponto $a$ (ou seja, $f'(a)$ existe), então $f$ *obrigatoriamente* é contínua no ponto $a$[cite: 132]. Intuitivamente, se você pode desenhar uma reta tangente única, o gráfico não pode ter uma quebra ali.
* **Continuidade NÃO implica Diferenciabilidade:** Uma função pode ser contínua em um ponto $a$, mas não ser diferenciável ali[cite: 138]. O exemplo clássico é a função valor absoluto $f(x) = |x|$ no ponto $x=0$. O gráfico tem um "bico" (ou "quina") em $x=0$. Ela é contínua ali, mas não há uma única reta tangente bem definida, então a derivada não existe em $x=0$[cite: 139, 142]. Outros exemplos incluem funções com tangentes verticais[cite: 142].

Em resumo: uma função que possui uma derivada é contínua, mas existem funções contínuas que não possuem derivada[cite: 142].

1. Análise de Taxas de Variação (Ex: Velocidade e Aceleração)

A derivada nos dá a taxa com que uma quantidade muda. Se temos a posição de um objeto em função do tempo,

````python
import sympy

# Definir o tempo como um símbolo
t_sym = sympy.symbols('t')

# Função de posição: s(t) = 4.9*t^2 + 10*t + 5  (exemplo de um objeto em movimento)
posicao_t = 4.9 * t_sym**2 + 10 * t_sym + 5
print(f"Função de Posição s(t): {posicao_t}")

# Velocidade v(t) é a primeira derivada da posição em relação ao tempo
velocidade_t = sympy.diff(posicao_t, t_sym)
print(f"Função de Velocidade v(t) = s'(t): {velocidade_t}")

# Aceleração a(t) é a primeira derivada da velocidade (ou segunda da posição)
aceleracao_t = sympy.diff(velocidade_t, t_sym)
print(f"Função de Aceleração a(t) = v'(t) = s''(t): {aceleracao_t}")

# Calcular velocidade e aceleração em um tempo específico, por exemplo, t = 2 segundos
tempo_especifico = 2
velocidade_no_tempo = velocidade_t.subs(t_sym, tempo_especifico)
aceleracao_no_tempo = aceleracao_t.subs(t_sym, tempo_especifico) # Deverá ser constante neste exemplo

print(f"\nEm t = {tempo_especifico}s:")
print(f"  Posição s({tempo_especifico}) = {posicao_t.subs(t_sym, tempo_especifico)} metros")
print(f"  Velocidade v({tempo_especifico}) = {velocidade_no_tempo} m/s")
print(f"  Aceleração a({tempo_especifico}) = {aceleracao_no_tempo} m/s²")
````

2. Otimização (Encontrar Máximos e Mínimos)

A derivada é zero em pontos de máximo ou mínimo locais de funções suaves.

````python
import sympy
import numpy as np
import matplotlib.pyplot as plt

x_sym = sympy.symbols('x')

# Função a ser otimizada: f(x) = x^3 - 6x^2 + 9x + 1
f_x = x_sym**3 - 6*x_sym**2 + 9*x_sym + 1
print(f"Função f(x): {f_x}")

# Primeira derivada para encontrar pontos críticos
f_prime_x = sympy.diff(f_x, x_sym)
print(f"Primeira Derivada f'(x): {f_prime_x}")

# Pontos críticos são onde f'(x) = 0
pontos_criticos_sym = sympy.solve(f_prime_x, x_sym)
print(f"Pontos Críticos (onde f'(x)=0): {pontos_criticos_sym}")

# Segunda derivada para classificar os pontos críticos
f_double_prime_x = sympy.diff(f_prime_x, x_sym)
print(f"Segunda Derivada f''(x): {f_double_prime_x}")

for p_critico in pontos_criticos_sym:
    valor_segunda_derivada = f_double_prime_x.subs(x_sym, p_critico)
    valor_f_no_ponto = f_x.subs(x_sym, p_critico)
    if valor_segunda_derivada > 0:
        print(f"  Ponto x={p_critico} é um MÍNIMO local (f({p_critico})={valor_f_no_ponto}), f''({p_critico})={valor_segunda_derivada}")
    elif valor_segunda_derivada < 0:
        print(f"  Ponto x={p_critico} é um MÁXIMO local (f({p_critico})={valor_f_no_ponto}), f''({p_critico})={valor_segunda_derivada}")
    else:
        print(f"  Ponto x={p_critico} (f({p_critico})={valor_f_no_ponto}). Teste da segunda derivada inconclusivo.")

# Visualização
f_lambdify = sympy.lambdify(x_sym, f_x, 'numpy')
x_vals = np.linspace(-1, 5, 400)
y_vals = f_lambdify(x_vals)

plt.figure(figsize=(8,6))
plt.plot(x_vals, y_vals, label=f"${sympy.latex(f_x)}$")
for p_critico in pontos_criticos_sym:
    plt.plot(p_critico, f_lambdify(p_critico), 'ro', label=f'Ponto crítico x={p_critico}')
plt.title("Otimização de Função")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
````

O que o código faz:

Define uma função, calcula sua primeira derivada para encontrar pontos críticos (onde a inclinação é zero) e usa a segunda derivada para classificar se são máximos ou mínimos. O gráfico ilustra isso.

3. Aproximações Lineares

A reta tangente (cuja inclinação é a derivada) é a melhor aproximação linear da função perto de um ponto.

````python
import sympy
import numpy as np
import matplotlib.pyplot as plt

x_sym = sympy.symbols('x')

# Função: f(x) = sqrt(x)
f_x_sqrt = sympy.sqrt(x_sym)
print(f"Função f(x): {f_x_sqrt}")

# Ponto 'a' para a aproximação
a = 4
f_a_sqrt = f_x_sqrt.subs(x_sym, a)

# Derivada de f(x)
f_prime_sqrt = sympy.diff(f_x_sqrt, x_sym)
f_prime_a_sqrt = f_prime_sqrt.subs(x_sym, a)
print(f"Derivada f'(x): {f_prime_sqrt}, em x={a}, f'({a})={f_prime_a_sqrt}")

# Aproximação linear L(x) = f(a) + f'(a)(x-a)
L_x = f_a_sqrt + f_prime_a_sqrt * (x_sym - a)
print(f"Aproximação Linear L(x) em torno de x={a}: {L_x}")

# Ponto a ser aproximado, ex: x_aprox = 4.1
x_aprox_val = 4.1
valor_real_aprox = f_x_sqrt.subs(x_sym, x_aprox_val).evalf() # .evalf() para valor numérico
valor_aproximado_L = L_x.subs(x_sym, x_aprox_val).evalf()

print(f"\nAproximando f({x_aprox_val}):")
print(f"  Valor Real f({x_aprox_val}) = {valor_real_aprox}")
print(f"  Valor Aproximado L({x_aprox_val}) = {valor_aproximado_L}")
print(f"  Erro da aproximação = {abs(valor_real_aprox - valor_aproximado_L)}")

# Visualização
f_lambdify_sqrt = sympy.lambdify(x_sym, f_x_sqrt, 'numpy')
L_lambdify = sympy.lambdify(x_sym, L_x, 'numpy')

x_plot_vals = np.linspace(0, 9, 400) # sqrt(x) definido para x>=0
y_f_vals = f_lambdify_sqrt(x_plot_vals)
y_L_vals = L_lambdify(x_plot_vals)

plt.figure(figsize=(8,6))
plt.plot(x_plot_vals, y_f_vals, label=f"${sympy.latex(f_x_sqrt)}$")
plt.plot(x_plot_vals, y_L_vals, '--', label=f"Aproximação Linear L(x) em $x={a}$")
plt.plot(a, f_lambdify_sqrt(a), 'ro', label=f'Ponto de tangência $({a}, {f_lambdify_sqrt(a)})$')
plt.plot(x_aprox_val, valor_real_aprox, 'go', label=f'Valor Real em $x={x_aprox_val}$')
plt.plot(x_aprox_val, valor_aproximado_L, 'mo', label=f'Valor Aproximado em $x={x_aprox_val}$')
plt.title("Aproximação Linear de $f(x)=\\sqrt{x}$")
plt.xlabel("x")
plt.ylabel("f(x), L(x)")
plt.legend()
plt.grid(True)
plt.ylim(0, 3.5)
plt.show()
````

4. Entender o Comportamento de Funções (Crescimento/Decrescimento)

O sinal da derivada indica se a função está crescendo ou decrescendo.

````python
import sympy
import numpy as np
import matplotlib.pyplot as plt

x_sym = sympy.symbols('x')

# Função: f(x) = x^3 - 3x
f_x_comp = x_sym**3 - 3*x_sym
print(f"Função f(x): {f_x_comp}")

# Primeira derivada
f_prime_comp = sympy.diff(f_x_comp, x_sym)
print(f"Primeira Derivada f'(x): {f_prime_comp}")

# Encontrar pontos críticos (onde f'(x) = 0) para definir intervalos
pontos_criticos_comp = sympy.solve(f_prime_comp, x_sym)
pontos_criticos_comp.sort() # Ordenar para facilitar análise de intervalo
print(f"Pontos Críticos: {pontos_criticos_comp}")

print("\nAnálise de Intervalos:")
# Intervalo 1: x < ponto_critico[0]
# Escolher um ponto de teste, ex: ponto_critico[0] - 1
if len(pontos_criticos_comp) >= 2:
    p_teste1 = pontos_criticos_comp[0] - 1
    sinal1 = f_prime_comp.subs(x_sym, p_teste1)
    print(f"  Para x < {pontos_criticos_comp[0]} (ex: x={p_teste1}), f'(x) = {sinal1}. Função {'Crescente' if sinal1 > 0 else 'Decrescente'}")

    # Intervalo 2: ponto_critico[0] < x < ponto_critico[1]
    p_teste2 = (pontos_criticos_comp[0] + pontos_criticos_comp[1]) / 2
    sinal2 = f_prime_comp.subs(x_sym, p_teste2)
    print(f"  Para {pontos_criticos_comp[0]} < x < {pontos_criticos_comp[1]} (ex: x={p_teste2}), f'(x) = {sinal2}. Função {'Crescente' if sinal2 > 0 else 'Decrescente'}")

    # Intervalo 3: x > ponto_critico[1]
    p_teste3 = pontos_criticos_comp[1] + 1
    sinal3 = f_prime_comp.subs(x_sym, p_teste3)
    print(f"  Para x > {pontos_criticos_comp[1]} (ex: x={p_teste3}), f'(x) = {sinal3}. Função {'Crescente' if sinal3 > 0 else 'Decrescente'}")
elif len(pontos_criticos_comp) == 1: # Caso de uma parábola ou similar com 1 ponto crítico
    p_teste1 = pontos_criticos_comp[0] - 1
    sinal1 = f_prime_comp.subs(x_sym, p_teste1)
    print(f"  Para x < {pontos_criticos_comp[0]} (ex: x={p_teste1}), f'(x) = {sinal1}. Função {'Crescente' if sinal1 > 0 else 'Decrescente'}")
    p_teste2 = pontos_criticos_comp[0] + 1
    sinal2 = f_prime_comp.subs(x_sym, p_teste2)
    print(f"  Para x > {pontos_criticos_comp[0]} (ex: x={p_teste2}), f'(x) = {sinal2}. Função {'Crescente' if sinal2 > 0 else 'Decrescente'}")


# Visualização
f_lambdify_comp = sympy.lambdify(x_sym, f_x_comp, 'numpy')
x_plot_cvals = np.linspace(-2.5, 2.5, 400)
y_f_cvals = f_lambdify_comp(x_plot_cvals)

plt.figure(figsize=(8,6))
plt.plot(x_plot_cvals, y_f_cvals, label=f"${sympy.latex(f_x_comp)}$")
for pc in pontos_criticos_comp:
    plt.axvline(pc, color='gray', linestyle='--', label=f'Ponto crítico x={pc}')
plt.title("Comportamento de $f(x)=x^3-3x$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
````

O que o código faz:

Calcula a derivada de f(x)=x 3 −3x, encontra os pontos onde a derivada é zero, e então analisa o sinal da derivada em intervalos definidos por esses pontos para determinar onde a função original é crescente ou decrescente.

5. Modelagem em Ciências (Ex: Decaimento Radioativo)
Muitas leis físicas são equações diferenciais. O decaimento radioativo diz que a taxa de decaimento de uma substância é proporcional à quantidade de substância presente:  
$\frac{dt}{dN}$ = −λN. Aqui, $\frac{dt}{dN}$ é a derivada.

````python
import sympy
import numpy as np
import matplotlib.pyplot as plt

# Símbolos
N_sym, t_sym, lambda_sym = sympy.symbols('N t lambda')

# A equação diferencial do decaimento radioativo
# dN/dt = -lambda * N
# Não vamos resolver a EDO aqui, mas mostrar como a derivada está presente.

# A solução conhecida (para N(0) = N0) é N(t) = N0 * exp(-lambda*t)
N0_val = 100  # Quantidade inicial
lambda_val = 0.1 # Constante de decaimento

# Função N(t)
N_t_decay = N0_val * sympy.exp(-lambda_val * t_sym)
print(f"Modelo de Decaimento N(t): {N_t_decay}")

# Derivada dN/dt (taxa de decaimento)
dN_dt = sympy.diff(N_t_decay, t_sym)
print(f"Taxa de Decaimento dN/dt: {dN_dt}")

# Verificar se dN/dt = -lambda * N(t)
print(f"Verificação: -lambda*N(t) = {-lambda_val * N_t_decay}")
# As duas últimas expressões devem ser iguais.

# Visualização
N_lambdify = sympy.lambdify(t_sym, N_t_decay, 'numpy')
t_plot_vals = np.linspace(0, 50, 200)
N_plot_vals = N_lambdify(t_plot_vals)

dN_dt_lambdify = sympy.lambdify(t_sym, dN_dt, 'numpy')
rate_plot_vals = dN_dt_lambdify(t_plot_vals)

fig, ax1 = plt.subplots(figsize=(10,6))

color = 'tab:blue'
ax1.set_xlabel('Tempo (t)')
ax1.set_ylabel('Quantidade N(t)', color=color)
ax1.plot(t_plot_vals, N_plot_vals, color=color, label='N(t) - Quantidade')
ax1.tick_params(axis='y', labelcolor=color)
ax1.grid(True, linestyle=':', alpha=0.7)

ax2 = ax1.twinx()  # Compartilha o mesmo eixo x
color = 'tab:red'
ax2.set_ylabel('Taxa de Decaimento dN/dt', color=color)
ax2.plot(t_plot_vals, rate_plot_vals, color=color, linestyle='--', label='$\\frac{dN}{dt}$ - Taxa de Decaimento')
ax2.tick_params(axis='y', labelcolor=color)

fig.suptitle("Modelo de Decaimento Radioativo", fontsize=15)
# Para adicionar legendas de ambos os eixos
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper right')

fig.tight_layout(rect=[0, 0, 1, 0.96]) # Ajuste para o suptitle
plt.show()
````

O que o código faz:

Define a função de decaimento radioativo, calcula sua derivada (que representa a taxa de decaimento) e plota ambas para mostrar como a taxa de decaimento (negativa) diminui à medida que a quantidade da substância diminui.

6. Ciência de Dados e Machine Learning (Um Passo da Descida do Gradiente)

A Descida do Gradiente usa a derivada para encontrar o mínimo de uma função de erro.

````python
import sympy

x_sym = sympy.symbols('x')

# Função de erro (ou custo) simples: f(x) = (x - 3)^2  (mínimo em x=3)
func_erro = (x_sym - 3)**2
print(f"Função de Erro f(x): {func_erro}")

# Derivada da função de erro
grad_erro = sympy.diff(func_erro, x_sym)
print(f"Gradiente (Derivada) f'(x): {grad_erro}")

# Parâmetros da Descida do Gradiente
x_atual = 0.0  # Ponto de partida
taxa_aprendizado = 0.1 # Alpha
num_iteracoes = 20

print(f"\nIniciando Descida do Gradiente em x = {x_atual}")
for i in range(num_iteracoes):
    grad_no_ponto = grad_erro.subs(x_sym, x_atual)
    x_novo = x_atual - taxa_aprendizado * grad_no_ponto
    print(f"Iteração {i+1}: x = {x_atual:.4f}, f(x) = {func_erro.subs(x_sym, x_atual):.4f}, grad = {grad_no_ponto:.4f} -> x_novo = {x_novo:.4f}")
    if abs(x_novo - x_atual) < 1e-4: # Critério de parada simples
        print("Convergência atingida.")
        break
    x_atual = x_novo

print(f"\nPonto de mínimo encontrado (aproximado): x = {x_atual:.4f}")
print(f"Valor da função no mínimo: f({x_atual:.4f}) = {func_erro.subs(x_sym, x_atual):.4f}")

# Para backpropagation, a ideia é similar, mas com muitas variáveis (pesos da rede)
# e a Regra da Cadeia é usada para calcular os gradientes eficientemente.
print("\nEm Backpropagation (conceitualmente):")
print("1. Calcula-se o erro da rede.")
print("2. A derivada desse erro em relação a cada peso (gradiente) é calculada usando a Regra da Cadeia.")
print("3. Os pesos são ajustados na direção oposta ao gradiente para minimizar o erro (similar à descida do gradiente).")
````

````python
import matplotlib.pyplot as plt
import numpy as np

# 1. Definir a função e sua derivada
def f(x):
    return x**2

def df(x): # Derivada de f(x)
    return 2*x

# 2. Ponto de tangência
a = 1
f_a = f(a)
df_a = df(a)

# 3. Valores de x para plotagem da função principal
x_vals = np.linspace(a - 2, a + 2, 400)
y_vals = f(x_vals)

# 4. Calcular os pontos da reta tangente
# y_tangente = f(a) + f'(a)*(x - a)
y_tangente = f_a + df_a * (x_vals - a)

# 5. Criar a figura e os eixos
plt.figure(figsize=(10, 8))

# Plotar a função principal f(x)
plt.plot(x_vals, y_vals, 'b-', label='$f(x) = x^2$', linewidth=2)

# Plotar a reta tangente
plt.plot(x_vals, y_tangente, 'r--', label=f'Reta Tangente em $x={a}$ ($y={df_a:.1f}x {f_a - df_a*a:+.1f}$)', linewidth=2)

# Ponto de tangência
plt.plot(a, f_a, 'ko', markersize=8, label=f'Ponto de tangência $({a}, {f_a})$')

# 6. Plotar retas secantes para diferentes valores de h
valores_h = [2, 1, 0.5, 0.1]
colors = ['green', 'purple', 'orange', 'cyan']

for i, h in enumerate(valores_h):
    # Ponto (a+h, f(a+h))
    x_h = a + h
    f_ah = f(x_h)

    # Inclinação da reta secante
    m_secante = (f_ah - f_a) / h
    # y_secante = f(a) + m_secante * (x - a)
    y_secante = f_a + m_secante * (x_vals - a)

    plt.plot(x_vals, y_secante, linestyle=':', color=colors[i],
             label=f'Reta Secante com $h={h}$ ($m_s \\approx {m_secante:.2f}$)', linewidth=1.5)
    # Marcar o segundo ponto da secante
    plt.plot(x_h, f_ah, 'o', color=colors[i], markersize=5)


# 7. Configurações do gráfico
plt.title(f'Reta Tangente e Retas Secantes para $f(x) = x^2$ em $x={a}$', fontsize=15)
plt.xlabel('$x$', fontsize=12)
plt.ylabel('$f(x)$', fontsize=12)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle=':', alpha=0.7)
plt.legend(fontsize=9, loc='upper left')
plt.ylim(min(y_vals.min(), y_tangente.min()) - 1, max(y_vals.max(), y_tangente.max()) + 1) # Ajuste de limites do eixo y
plt.xlim(x_vals.min(), x_vals.max()) # Ajuste de limites do eixo x

plt.show()
````
